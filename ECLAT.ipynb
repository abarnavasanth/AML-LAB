```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("C:\\Users\\Abarna V\\Desktop\\End semester AML Laboratory (Dataset)\\ECLAT\\groceries2.csv")
df
```

```python
trans = []
for i in range(0, 6105):
    trans.append([str(df.values[i,j]) for j in range(0, 4)])
```

```python
print(trans)
```

```python
from apyori import apriori
rules = apriori(transactions = trans, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 10)
```

```python
print(rules)
```

```python
results = list(rules)
results
```

```python
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    return list(zip(lhs, rhs, supports))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Product 1', 'Product 2', 'support'])
```

```python
print(results)
```

```python
resultsinDataFrame.nlargest(n = 10, columns = 'support')
```
